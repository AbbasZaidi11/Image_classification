```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
import pickle
```

```python
data=[]
IMG_SIZE = 100 
```

```python

DIR = '/kaggle/input/a-large-scale-fish-dataset/Fish_Dataset/Fish_Dataset'

for dirname, subdirs, filenames in os.walk(DIR):
     for subdir in subdirs:
        if not subdir.endswith("GT"):
            subdirectory_path = os.path.join(dirname, subdir)
            for filename in os.listdir(subdirectory_path):
                if filename.endswith('.png'):  # If filename ends with .png
                    image_path = os.path.join(subdirectory_path, filename)
                    img_arr = cv2.imread(image_path)
                    img_arr = cv2.resize(img_arr, (IMG_SIZE, IMG_SIZE))
                    data.append([img_arr, subdir])

```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
x = []
y = []

for features,labels in data:
    x.append(features)
    y.append(labels)
```

```python
X = np.array(x)
Y = np.array(y)
# converting the lists into numpy arrays
```
